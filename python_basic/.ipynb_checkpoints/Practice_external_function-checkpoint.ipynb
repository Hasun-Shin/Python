```python
import pickle
class NowData:
    def __init__(self):
        self.x = 0
        self.y = 0
    def move(self, x, y):
        self.x = x
        self.y = y
    def get_location(self):
        return self.x, self.y
data = NowData()
data.move(100, 200)

file = open('data.bin', 'wb')
pickle.dump(data,file)
file.close
# 코드 작성
```

```python
import pickle
file = open('data.bin','rb')
data = pickle.load(file)
data.get_location()
```
