## 정규 표현식 (Regular Expressions)<br>
 ● 특정한 규칙을 가진 문자열의 집합을 표현할때 사용하는 언어<br>
 ● 여러 텍스트 편집기와 프로그래밍 언어에서 지원<br>
 ● 정규 표현식이 유용하게 사용되는 경우
 - 영어와 숫자로만 작성되어야 하는 회원 아이디 검사
 - 한 글자가 특정 횟수 이상 반복되는 경우 검사
 - 주민등록번호, 이메일, 휴대폰 번호 등의 형식 검사
 - 비밀번호 생성 시 특수문자 포함 여부 검사
 
 
 
 ● 정규식에서 사용되는 메타 문자 (특별한 용도로 사용되는 문자)<br>
 ● 문자 클래스 (문자 매치)<br>
 - 대괄호([ ]) 사이에 문자 또는 문자열을 입력
 - 두 문자 사이에 하이픈(-)을 사용하면 범위(From - To)를 의미<Br>
. ^ $ * + ? { } [ ] \ | (


● 문자 클래스 (문자 매치)
 - 대괄호([ ]) 사이에 문자 또는 문자열을 입력
 - 두 문자 사이에 하이픈(-)을 사용하면 범위(From - To)를 의미<Br>

- 예시 : [abc] # a 또는 b 또는 c <br>
     a # a가 있으므로 매치<br>
     black # b 또는 a가 있으므로 매치<br>
     duke # 일치하는 문자가 없음<br>
    [a-z] # a 부터 z 까지 모든 알파벳 문자<br>
    [0-9] # 0 부터 9까지 모든 숫자<br>
    [0-3a-dA-D] # 0 ~ 3, a ~ d, A ~ D

```python

import re
pat = re.compile('[abc]')
s = 'python'
print(pat.match(s)) # 맞지 않아 None 이라고 뜸 
```

```python

import re
pat = re.compile('[abc]')
s = 'apython'
if pat.match(s):
    print(True)
else:
    print(False)
```

```python

import re
pat = re.compile('[a-z0-9]') #범위를 나타냄.
s = 'apython'
if pat.match(s):
    print(True)
else:
    print(False)
```

[^abc] # a 또는 b 또는 c 가 아닌 문자<br>
[^0-9] # 0 부터 9까지 모든 숫자를 제외한 문자<br>
[^0-3a-dA-D] # 0 ~ 3, a ~ d, A ~ D 를 제외한 문자




```python
import re
pat = re.compile('[ㄱ-힇]')
s = 'apython'
if pat.match(s):
    print(True)
else:
    print(False)
```

```python
import re
pat = re.compile('^a.*[xz]$') #a로 시작하고 중간에 뭐가와도 상관없고 끝나는것만 x나 z 이면 됨. 
s = 'adferfehgz'
if pat.match(s):
    print(True)
else:
    print(False)
```

```python
import re
text = '791111-1234567'


regex = '[0-9]{6}-([1-4]{1}\d{6}$)' # 소괄호. 그룹화. 
pattern = re.compile(regex)
res = pattern.match(text)
print(res)
print(res.group(1))
```

```python
import re
text = '안녕하세요. 저는 OOO에서 태어났고... 열심히 하겠습니다. 감사합니다.'
regex = '^안녕하세요.*열심히.*감사합니다.$'
pattern = re.compile(regex)
res = pattern.match(text)
print(res)


#  '안녕하세요' 라는 문자열로 시작
#  - 내용 중 '열심히' 라는 문자열 반드시 포함
#  - '감사합니다.' 라는 문자열로 종료
```

```python
import re
text = '이름:꼬랩 , 전화번호:1234'
regex = '이름:(.+), 전화번호:(\d+)'
pattern = re.compile(regex)
res = pattern.match(text)
print(res.group(1))
print(res.group(2))
```

```python
import re
text = '12ab1234'
regex = '[0-9]{2}[a-z]{2}[0-9]{4}'
pattern = re.compile(regex)
res = pattern.match(text)
print(res)

# 숫자 2자리 + 영어 소문자 2자리 + 숫자 4자리로 구성되었는지 확인하는 정규식 만들기
```

```python
import re
text = '가1234d가'
regex = '^[a-zA-Z0-9]+$'
pattern = re.compile(regex)
res = pattern.match(text)
print(res)
```

```python
import re
text = '1번 a@a.com 2번 b@b.org 3번 c@c.net'
regex = '\w+@\w+[.](com|net)'
pattern = re.compile(regex)
res = pattern.findall(text)
print(res)


#- 숫자 3자리-숫자 2자리-숫자8자리
```

finditer : 정규식과 매치되는 모든 문자열을 iterator 객체로 반환한다.
<br>
com 또는 net 이 있는 문자열을 반환한다. 

```python
import re
text = '1번 a@a.com 2번 b@b.org 3번 c@c.net'
regex = '\w+@\w+[.](com|net)'
pattern = re.compile(regex)
res = pattern.finditer(text)
for it in res:
    print(it)
```

```python
import re
text = '010-8478-8181'
regex = '(\d{3})-\d{3,4}-(\d{4})'
pattern = re.compile(regex)
change = '\g<1>-****-\g<2>'
res = pattern.sub(change, text)
print(res)



# 전화번호 중 중간자리 숫자 마킹 처리
```

```python
import re
user_list = ['ggoreb@naver.com', 'seorab@gmail.com', 'human@ggoreb.com']
regex = '(\w+@)\w+([.]com)'
pattern = re.compile(regex)
change = '\g<1>*****\g<2>'
for user in user_list:
    res = pattern.sub(change, user)
    print(res)
    
#     이메일의 도메인 앞부분 마킹 처리
```
