## 인터파크 여행상품 '달랏' 관련 목록 나타내기

```python
def interpark():
    
    from selenium.webdriver.common.by import By
    from selenium.webdriver.support.ui import WebDriverWait
    from selenium.webdriver.support import expected_conditions as E
    from selenium import webdriver as wd
    import time

    driver = wd.Chrome(executable_path='chromedriver.exe')
    driver.get('http://tour.interpark.com')

    driver.find_element_by_tag_name('#SearchGNBText').send_keys('달랏')
    driver.find_element_by_css_selector('#dHead > div.tour-header > div.search-area > form > fieldset > button.search-btn').click()
    driver.find_element_by_css_selector('body > div.container > div > div > div.panelZone > div.oTravelBox > ul > li.moreBtnWrap > button').click()
    
    
    
    
    for page in range(1,6):
        driver.find_element_by_xpath("/html/body/div[3]/div/div/div[5]/div[4]/ul/li[%s]" %page).click()
        
        
        
        time.sleep(2)
        
        for number in range(1,11):
            
            imgs = driver.find_elements_by_css_selector('body > div.container > div > div > div.panelZone > div.oTravelBox > ul > li:nth-child(%s) > a > img' %number)
            for img in imgs:
                img_val = img.get_attribute("src")
                print(img_val)
        
            links = driver.find_elements_by_css_selector('body > div.container > div > div > div.panelZone > div.oTravelBox > ul > li:nth-child(%s) > div > div.title-row > div:nth-child(1) > h5' %number)
            for link in links:
                link_val = link.get_attribute("onclick")
                print(link_val)

            titles = driver.find_elements_by_css_selector('body > div.container > div > div > div.panelZone > div.oTravelBox > ul > li:nth-child(%s) > div > div.title-row > div:nth-child(1) > h5' %number)
            for title in titles:

                print(title.text)

            extras = driver.find_elements_by_css_selector('body > div.container > div > div > div.panelZone > div.oTravelBox > ul > li:nth-child(%s) > div > div.title-row > div:nth-child(1) > p' %number)
            for extra in extras:
                print(extra.text)

            costs = driver.find_elements_by_css_selector('body > div.container > div > div > div.panelZone > div.oTravelBox > ul > li:nth-child(%s) > div > div.title-row > div:nth-child(2) > strong' %number)
            for cost in costs:
                print(cost.text)

            dates = driver.find_elements_by_css_selector('body > div.container > div > div > div.panelZone > div.oTravelBox > ul > li:nth-child(%s) > div > div.info-row > div:nth-child(1) > p:nth-child(1)' %number)
            for date in dates:
                print(date.text)

            departures = driver.find_elements_by_css_selector('body > div.container > div > div > div.panelZone > div.oTravelBox > ul > li:nth-child(%s) > div > div.info-row > div:nth-child(1) > p:nth-child(2)' %number)
            for departure in departures:
                print(departure.text)

            rates= driver.find_elements_by_css_selector('body > div.container > div > div > div.panelZone > div.oTravelBox > ul > li:nth-child(%s) > div > div.info-row > div:nth-child(2) > p:nth-child(2)' %number)
            for rate in rates:
                print(rate.text)

            comments = driver.find_elements_by_css_selector('body > div.container > div > div > div.panelZone > div.oTravelBox > ul > li:nth-child(%s) > div > div.info-row > div:nth-child(2) > p:nth-child(3)' %number)
            for comment in comments:
                print(comment.text)

            print("========================================================")
            

        

        
```

```python
interpark()
```
