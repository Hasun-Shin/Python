```python
import pymysql

conn = None #전역변수로 conn 변수 생성함. 

try:
    conn = pymysql.connect(
        host ='localhost', user='root', password=비밀번호,
        db='python', charset='utf8') #데이터베이스 접속

    with conn.cursor() as cursor: #Cursor 객체 생성

        sql = '''CREATE TABLE mysql (
                    id INTEGER PRIMARY KEY AUTO_INCREMENT
                    , title VARCHAR(100), content VARCHAR(100))'''

        cursor.execute(sql) #sql 실행

        conn.commit() # 저장


except pymysql.InternalError as e:
        
    print(e)
finally:
    if conn:
        conn.close() # try 내의 전역변수 
```

```python
#  article  테이블 만들기


import pymysql


conn = pymysql.connect(
host ='localhost', user='root', password=비밀번호,
db='python', charset='utf8') #데이터베이스 접속

cursor = conn.cursor() #Cursor 객체 생성

sql = '''CREATE TABLE article (
                    ID INTEGER PRIMARY KEY AUTO_INCREMENT
                    , TITLE VARCHAR(30), CONTENT VARCHAR(30), WRITER VARCHAR(30),
                    HIT INTEGER, C_DATE DATETIME)'''

           

cursor.execute(sql) #sql 실행

conn.commit() 


cursor.close()
conn.close() # try 내의 전역변수 
```

```python
#  article  insert


import pymysql


conn = pymysql.connect(
host ='localhost', user='root', password='1234',
db='python', charset='utf8') #데이터베이스 접속

cursor = conn.cursor() #Cursor 객체 생성

# sql = '''UPDATE ARTICLE SET
#         TITLE = %s, CONTENT = %s, HIT = %s
#         WHERE ID - %s'''

# 데이터 100개 입력
sql = '''INSERT INTO ARTICLE
 (ID, TITLE, CONTENT, WRITER, HIT, C_DATE)
 VALUES (NULL, %s, %s, %s, 0, NOW())'''
for i in range(1, 101):
    cursor.execute(sql, ('제목' + str(i), '내용' + str(i), '파이썬'))           

#cursor.execute(sql,('제목수정','내용수정',100, 1)) #sql 실행

conn.commit() 


cursor.close()
conn.close() # try 내의 전역변수 
```

```python
import pymysql

page_num = input('원하는 페이지 번호를 입력해주세요.')
#print(type(page_num)) # str 문자열 타입임.
page_num = int(page_num)
end_num = page_num *10
start_num = end_num - 9


conn = pymysql.connect(
host ='localhost', user='root', password='1234',
db='python', charset='utf8') #데이터베이스 접속

cursor = conn.cursor(pymysql.cursors.DictCursor) #Cursor 객체 생성

sql = '''SELECT *
            FROM article
            LIMIT %s, 5'''

cursor.execute(sql,(start_num-1)) #sql 실행

result = cursor.fetchall() #fetchall() 모두 조회 

print(len(result))


for row in result:
    print(row['ID'], row['TITLE'], row['CONTENT'], row['WRITER'],row['HIT'],row['C_DATE'])


conn.commit() 


cursor.close()
conn.close() # try 내의 전역변수
```

```python
import pymysql


conn = pymysql.connect(
host ='localhost', user='root', password='1234',
db='mydb', charset='utf8') #데이터베이스 접속

cursor = conn.cursor(pymysql.cursors.DictCursor) #Cursor 객체 생성

sql = '''SELECT id, year, area1, count(*) as cnt
            FROM accident
            GROUP BY AREA1, YEAR'''

cursor.execute(sql) #sql 실행

result = cursor.fetchall() #fetchall() 모두 조회 
for row in result:
    print(row['area1'], row['year'], row['cnt'])


conn.commit() 


cursor.close()
conn.close() # try 내의 전역변수
```

```python
import pymysql
import re 



conn = pymysql.connect(
host ='localhost', user='root', password='1234',
db='mydb', charset='utf8') #데이터베이스 접속

cursor = conn.cursor(pymysql.cursors.DictCursor) #Cursor 객체 생성

sql = '''SELECT *
            FROM INFO
            '''

cursor.execute(sql) #sql 실행

result = cursor.fetchall() #fetchall() 모두 조회 

print(len(result))
for row in result:
    #print(row['ID'], row['NAME'], row['PERSON_NUMBER'])
    a = row['ID']
    b = row['NAME']
    c = row['PERSON_NUMBER']
    text = c 
    patturn = re.compile('(\d{6})-\d{7}')
    change = '\g<1>-*******'
    res = patturn.sub(change,text)
    print(a,b,res)


conn.commit() 


cursor.close()
conn.close() # try 내의 전역변수
```
